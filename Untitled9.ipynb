```python
import matplotlib.pyplot as plt
import numpy as np

# محدوده مقادیر امگا (محور افقی)
omega = np.linspace(-3 * np.pi, 3 * np.pi, 1000)
# این خط یک آرایه از 1000 نقطه را بین -3π و 3π (برای محور امگا) ایجاد می‌کند.
# محدوده را گسترده‌تر می‌کنیم تا شکل نمودار بهتر نمایش داده شود.

# تعریف تابع W(e^jω) (شبیه به تصویر)
def W(omega):
    # این تابع مقادیر W(e^jω) را بر اساس امگا محاسبه می‌کند.
    result = np.zeros_like(omega)
    # یک آرایه با طول آرایه امگا ایجاد می‌کنیم که با صفر مقداردهی شده است.
    for i, w in enumerate(omega):
        if abs(w) <= 0.5:
            result[i] = 1 / (1 + (w)**2)
        elif abs(w) <= 1:
            result[i] = 1 / (1 + (w)**2) * (1 - (abs(w) - 0.5) / 0.5)
        else:
            result[i] = 0
    return result

# محاسبه مقادیر W(e^jω)
w_values = W(omega)
# این خط مقادیر W(e^jω) را برای هر نقطه در آرایه امگا محاسبه می‌کند.

# ایجاد نمودار
plt.figure(figsize=(10, 5))
# این خط یک شکل جدید با اندازه 10x5 اینچ ایجاد می‌کند.
plt.plot(omega, w_values, color='black')
# این خط نمودار W(e^jω) را بر حسب امگا رسم می‌کند.

# اضافه کردن خط عمودی در امگا=0
plt.axvline(x=0, color='black')
# این خط یک خط عمودی در امگا = 0 اضافه می‌کند.

# اضافه کردن برچسب‌ها و عنوان
plt.xlabel('ω')
# این خط برچسب محور افقی را اضافه می‌کند.
plt.ylabel('W(e^jω)')
# این خط برچسب محور عمودی را اضافه می‌کند.
plt.title('(d)')
# این خط عنوان نمودار را اضافه می‌کند.

# تنظیمات محورها
plt.xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
# این خط مقادیر محور افقی را به [-2π, -π, 0, π, 2π] تنظیم می‌کند.
plt.yticks([])
# این خط مقادیر محور عمودی را حذف می‌کند.

# نمایش نمودار
plt.show()
# این خط نمودار را نمایش می‌دهد.
```

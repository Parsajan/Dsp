```python
import matplotlib.pyplot as plt
import numpy as np

# محدوده مقادیر امگا (محور افقی)
omega = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
# این خط یک آرایه از 1000 نقطه را بین -4π و 4π (برای محور امگا) ایجاد می‌کند.
# محدوده را گسترده‌تر می‌کنیم تا شکل نمودار بهتر نمایش داده شود.

# تعریف تابع Haa(jΩ) (شبیه به تصویر)
def Haa(omega, T):
    # این تابع مقادیر Haa(jΩ) را بر اساس امگا و T محاسبه می‌کند.
    # T دوره تناوب است.
    result = np.zeros_like(omega)
    # یک آرایه با طول آرایه امگا ایجاد می‌کنیم که با صفر مقداردهی شده است.
    for i, w in enumerate(omega):
        if -np.pi / T <= w <= np.pi / T:
            result[i] = 1
        elif -1.2 * np.pi / T <= w <= -np.pi / T:
            result[i] = 1 - (T * abs(w) - np.pi) / (0.2 * np.pi)
        elif np.pi / T <= w <= 1.2 * np.pi / T:
            result[i] = 1 - (T * abs(w) - np.pi) / (0.2 * np.pi)
        else:
            result[i] = 0
    return result

# مقدار دوره تناوب T را تعیین می‌کنیم
T = 1

# محاسبه مقادیر Haa(jΩ)
haa_values = Haa(omega, T)
# این خط مقادیر Haa(jΩ) را برای هر نقطه در آرایه امگا محاسبه می‌کند.

# ایجاد نمودار
plt.figure(figsize=(10, 5))
# این خط یک شکل جدید با اندازه 10x5 اینچ ایجاد می‌کند.
plt.plot(omega, haa_values, color='black')
# این خط نمودار Haa(jΩ) را بر حسب امگا رسم می‌کند.

# اضافه کردن خطوط عمودی در π/T و -π/T
plt.axvline(x=np.pi / T, color='gray', linestyle='--')
# این خط یک خط عمودی در امگا = π/T اضافه می‌کند.
plt.axvline(x=-np.pi / T, color='gray', linestyle='--')
# این خط یک خط عمودی در امگا = -π/T اضافه می‌کند.
# 'linestyle='--' خطوط را به صورت خط‌چین نمایش می‌دهد.

# اضافه کردن خط عمودی در امگا=0
plt.axvline(x=0, color='black')
# این خط یک خط عمودی در امگا = 0 اضافه می‌کند.

# اضافه کردن برچسب‌ها و عنوان
plt.xlabel('Ω')
# این خط برچسب محور افقی را اضافه می‌کند.
plt.ylabel('Haa(jΩ)')
# این خط برچسب محور عمودی را اضافه می‌کند.
plt.title('(b)')
# این خط عنوان نمودار را اضافه می‌کند.

# اضافه کردن متن π/T و -π/T
plt.text(np.pi / T + 0.2, 0.1, r'$\frac{\pi}{T}$', fontsize=12)
# این خط متن 'π/T' را در مختصات (π/T + 0.2, 0.1) اضافه می‌کند.
plt.text(-np.pi / T - 0.5, 0.1, r'$-\frac{\pi}{T}$', fontsize=12)
# این خط متن '-π/T' را در مختصات (-π/T - 0.5, 0.1) اضافه می‌کند.

# تنظیمات محورها
plt.xticks([-np.pi / T, 0, np.pi / T])
# این خط مقادیر محور افقی را به [-π/T, 0, π/T] تنظیم می‌کند.
plt.yticks([0, 1])
# این خط مقادیر محور عمودی را به [0, 1] تنظیم می‌کند.

# نمایش نمودار
plt.show()
# این خط نمودار را نمایش می‌دهد.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# توضیحات:
# این نوت‌بوک Google Colab مراحل پردازش سیگنال را نشان می‌دهد، از جمله:
# 1. تولید سیگنال
# 2. اضافه کردن نویز
# 3. پنجره‌بندی (Windowing)
# 4. محاسبه تبدیل فوریه گسسته (DFT)
# 5. نمایش نتایج

# 1. تولید سیگنال x[n]
N = 100  # طول سیگنال
n = np.arange(N)
x = np.sin(2 * np.pi * 0.1 * n)  # سیگنال نمونه سینوسی

# 2. اضافه کردن نویز
noise = 0.5 * np.random.randn(N)  # نویز تصادفی گاوسی
x_noisy = x + noise

# 3. پنجره‌بندی (Windowing)
window = np.hamming(N)  # پنجره همینگ
v = x_noisy * window

# 4. محاسبه DFT
V = np.fft.fft(v)
f = np.fft.fftfreq(N)  # فرکانس‌ها

# 5. نمایش نتایج
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(n, x)
plt.title('x[n] (سیگنال اصلی)')
plt.xlabel('نمونه')
plt.ylabel('دامنه')

plt.subplot(2, 2, 2)
plt.plot(n, x_noisy)
plt.title('x[n] + نویز')
plt.xlabel('نمونه')
plt.ylabel('دامنه')

plt.subplot(2, 2, 3)
plt.plot(n, v)
plt.title('v[n] = w[n]x[n] (سیگنال پنجره‌بندی شده)')
plt.xlabel('نمونه')
plt.ylabel('دامنه')

plt.subplot(2, 2, 4)
plt.plot(f, np.abs(V))
plt.title('DFT(v[n]) (طیف فرکانسی)')
plt.xlabel('فرکانس نرمالیزه')
plt.ylabel('قدر مطلق')

plt.tight_layout()
plt.show()

# توضیحات تکمیلی:
# - در این نوت‌بوک، یک سیگنال سینوسی ساده به عنوان سیگنال اصلی x[n] تولید شده است.
# - سپس، نویز تصادفی گاوسی به سیگنال اصلی اضافه شده تا سیگنال نویزی x_noisy تولید شود.
# - برای کاهش اثرات نویز و بهبود تحلیل فرکانسی، سیگنال نویزی با یک پنجره همینگ ضرب شده است.
# - تبدیل فوریه گسسته (DFT) سیگنال پنجره‌بندی شده v[n] محاسبه شده تا طیف فرکانسی آن بدست آید.
# - نمودارها نشان می‌دهند که چگونه نویز به سیگنال اضافه می‌شود، چگونه پنجره‌بندی بر سیگنال تأثیر می‌گذارد و چگونه DFT طیف فرکانسی سیگنال را نشان می‌دهد.
# - شما می‌توانید پارامترهای مختلف مانند طول سیگنال، سطح نویز و نوع پنجره را تغییر دهید تا تأثیر آنها را بر نتایج مشاهده کنید.
```

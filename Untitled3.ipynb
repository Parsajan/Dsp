```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   x[n]
N = 100  #  
n = np.arange(N)
x = np.sin(2 * np.pi * 0.1 * n)  #   

# 2.   
noise = 0.5 * np.random.randn(N)  #   
x_noisy = x + noise

# 3.  (Windowing)
window = np.hamming(N)  #  
v = x_noisy * window

# 4.  DFT
V = np.fft.fft(v)
f = np.fft.fftfreq(N)  # 

# 5.  
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(n, x)
plt.title('x[n]')

plt.subplot(2, 2, 2)
plt.plot(n, x_noisy)
plt.title('x[n] + ')

plt.subplot(2, 2, 3)
plt.plot(n, v)
plt.title('v[n] = w[n]x[n]')

plt.subplot(2, 2, 4)
plt.plot(f, np.abs(V))
plt.title('DFT(v[n])')

plt.tight_layout()
plt.show()
```

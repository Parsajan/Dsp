```python
import matplotlib.pyplot as plt
import numpy as np

# محدوده مقادیر امگا (محور افقی)
omega = np.linspace(-10, 10, 400)
# این خط یک آرایه از 400 نقطه را بین -10 و 10 (برای محور امگا) ایجاد می‌کند.

# تعریف تابع Sc(jΩ) (شبیه به تصویر)
def Sc(omega):
    # این تابع مقادیر Sc(jΩ) را بر اساس امگا محاسبه می‌کند.
    # این تابع به گونه‌ای طراحی شده که شبیه به شکل نمودار در تصویر باشد.
    # تابع از سه بخش تشکیل شده است:
    # 1. یک تابع پایه (1 / (1 + omega**2)) که شکل کلی نمودار را ایجاد می‌کند.
    # 2. یک تابع گاوسی در اطراف Ω0 (0.5 * np.exp(-(omega - 2)**2))
    # 3. یک تابع گاوسی در اطراف -Ω0 (0.5 * np.exp(-(omega + 2)**2))
    return 1 / (1 + omega**2) + 0.5 * np.exp(-(omega - 2)**2) + 0.5 * np.exp(-(omega + 2)**2)

# محاسبه مقادیر Sc(jΩ)
sc_values = Sc(omega)
# این خط مقادیر Sc(jΩ) را برای هر نقطه در آرایه امگا محاسبه می‌کند.

# ایجاد نمودار
plt.figure(figsize=(10, 5))
# این خط یک شکل جدید با اندازه 10x5 اینچ ایجاد می‌کند.
plt.plot(omega, sc_values, color='black')
# این خط نمودار Sc(jΩ) را بر حسب امگا رسم می‌کند.

# اضافه کردن خطوط عمودی در Ω0 و -Ω0
plt.axvline(x=2, color='gray', linestyle='--')
# این خط یک خط عمودی در امگا = 2 (Ω0) اضافه می‌کند.
plt.axvline(x=-2, color='gray', linestyle='--')
# این خط یک خط عمودی در امگا = -2 (-Ω0) اضافه می‌کند.
# 'linestyle='--' خطوط را به صورت خط‌چین نمایش می‌دهد.

# اضافه کردن خط عمودی در امگا=0
plt.axvline(x=0, color='black')
# این خط یک خط عمودی در امگا = 0 اضافه می‌کند.

# اضافه کردن برچسب‌ها و عنوان
plt.xlabel('Ω')
# این خط برچسب محور افقی را اضافه می‌کند.
plt.ylabel('Sc(jΩ)')
# این خط برچسب محور عمودی را اضافه می‌کند.
plt.title('(a)')
# این خط عنوان نمودار را اضافه می‌کند.

# اضافه کردن متن Ω0 و -Ω0
plt.text(2.2, 0.1, 'Ω0', fontsize=12)
# این خط متن 'Ω0' را در مختصات (2.2, 0.1) اضافه می‌کند.
plt.text(-2.8, 0.1, '-Ω0', fontsize=12)
# این خط متن '-Ω0' را در مختصات (-2.8, 0.1) اضافه می‌کند.

# تنظیمات محورها
plt.xticks([-2, 0, 2])
# این خط مقادیر محور افقی را به [-2, 0, 2] تنظیم می‌کند.
plt.yticks([])
# این خط مقادیر محور عمودی را حذف می‌کند.

# نمایش نمودار
plt.show()
# این خط نمودار را نمایش می‌دهد.
```
